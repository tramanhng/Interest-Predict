```python
# Encoder cho các cột sub_grade, grade

list = ['sub_grade', 'grade']
for i in list:
    df[i] = LabelEncoder().fit_transform(df[i])
```

```python
df.head(6)
```

### Mô hình Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
feature_cols = ['grade','sub_grade']

x = df[feature_cols]
y = df['interest_rate']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state=0)
```

```python
lin_reg = LinearRegression()
lin_reg.fit(x_train, y_train)
```

```python
y_predict = lin_reg.predict(x_test)
y_predict
```

```python
print('The full R-square is : ' +str((lin_reg.score(x,y)*100).round(5)) + '%')
print('The train R-square is : ' +str((lin_reg.score(x_train, y_train)*100).round(5)) + '%')
print('The test R-square is : ' +str((lin_reg.score(x_test, y_test)*100).round(5)) + '%')
```

### Mô hình Polynomial Regression

```python
# The mean squared error
print('Mean squared error: %.2f'
      % mean_squared_error(y_poly_test, y_poly_predict))
# The coefficient of determination
print('Coefficient of determination: %.2f'
      % r2_score(y_poly_test, y_poly_predict))
```
